### 12.1 Overview
Wordclouds are a way of visualizing the words in your document(s). In general, a wordcloud function will count the frequency of the words in your text and create a graphic where the words are randomly place in a pane with the size and color of the words dependent on the frequency of the word.

Wordclouds are amazingly simple to do, but you can spend lots of time playing with them.

Let's see how to generate a wordcloud.

### 12.2 Word Cloud Example

```python
with open("emma_chapter_one.txt") as f:
    text = f.read()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wordcloud = WordCloud().generate(text)
plt.figure()
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```

If we want to change a feature, like the background color, we can use `ImageColorGenerator`

```python
from wordcloud import ImageColorGenerator

wordcloud = WordCloud(background_color="white").generate(text)
plt.figure
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```

We may want to reduce the number of words that it displays.  We can do that with `max_words`.

```python
from wordcloud import ImageColorGenerator

wordcloud = WordCloud(max_words = 25, background_color="white").generate(text)
plt.figure
plt.axis('off')
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()
```

We also can tweak the font size by using `max_font_size`

```python
wordcloud = WordCloud(max_font_size=40, max_words=30, background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

Finally, `wordcloud` has its own stopwords.  We can use those or use a set of our own.

```python
from wordcloud import STOPWORDS
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords = stopwords, max_font_size=50, max_words=30, background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
